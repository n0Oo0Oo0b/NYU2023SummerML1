## Calculate the Mean Squared Error (MSE) and Mean Absolute Error (MAE) to determine goodness of fit

### Reminder :

Given :
- a dataset : $(x_i, y_i)$, $i = 1, 2, 3, ..., N$
- a model : $\hat{y} = w_1x + w_0$

We can compute the following two error functions :
- Mean Squared Error: $\displaystyle MSE = \frac{1}{N}\sum_{i=1}^N || y_i - \hat{y_i}||^2$
- Mean Absolute Error: $\displaystyle MAE = \frac{1}{N}\sum_{i=1}^N |y_i - \hat{y_i}|$

```python
## To-do
## Calculate MSE and MAE
for n, (w1, w0) in enumerate(weights, start=1):
    y_hat = x*w1 + w0
    mse = np.mean((y - y_hat) ** 2)
    mae = np.mean(abs(y - y_hat))
    print(f"Weights {n}: {mse = :.3}, {mae = :.3}")
```

```python
# Reset x and y
y = df['PRICE'].values
x = df['RM'].values

# Calculate optimal weights
x_new = np.pad(x[:, np.newaxis], [(0, 0), (1, 0)], constant_values=1)
optimal_weights = np.linalg.inv(x_new.T @ x_new) @ x_new.T @ y
```

```python
# Loss
y_hat = np.dot(x_new, optimal_weights)

mse = np.mean((y - y_hat) ** 2)
mae = np.mean(abs(y - y_hat))
print(f"Optimal weights: {mse = :.3}, {mae = :.3}")
```

```python
# Plot data points
plt.plot(x,y,'o')

# Plot line
w0, w1 = optimal_weights
xplt = np.linspace(3,9,100)
yplt = w1*xplt + w0
plt.plot(xplt, yplt, '-', linewidth=3)
plt.xlabel('Average number of rooms in a region')
plt.ylabel('Price')
plt.grid()
```
